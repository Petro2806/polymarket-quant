task 1.1

```python
df = pd.read_csv(filename)
df['bid_ask_spread'] = df['best_ask'] - df['best_bid']
```

Bid–ask spreads reflect the image of the market. Things like liquidity, competitiveness, execution risks and expected edge.
Detailed explanation for tight and wide spreads are below.  


```python
tightest_spreads = df.nsmallest(3, 'spread')[['question', 'bid_ask_spread']]
display(tightest_spreads)
```

Tight spreads is probably because of many participants that actively quoting. And probably big liquifity. 
which leads to deeper books, and overall low execution risk. 
So a lot of MM and competition among them is high, so the expected edge is relatively small. 
Expected edge is small, but at the same time low risks

```python
widest_spreads = df.nlargest(3, 'spread')[['question', 'bid_ask_spread']]
display(widest_spreads)
```


Wide spreads, on the other hand, can suggest as that liquidity is quite thin. They often appear in markets with fewer active market makers, lower depth, or higher volatility. In these environments, competition is weaker and a market maker can charge a larger spread, increasing the potential edge — but at the cost of higher inventory risk and a greater chance of adverse selection.

Flow imbalance matters to MMs because it reveals the direction of aggressive order flow. 
When taker-buy flow consistently dominates, MMs ask is more likely to be hit just before the price moves up, so apeers losses on filled quotes. Similarly, for taker-sell. 

If MM keeps providing liquidity against the direction of the flow, they accumulate inventory against the trend, which amplifies inventory risk. By monitoring flow imbalance, a MM can widen spreads, change quotes, or reduce size on ask/bid (up to situation) to better manage risk.
